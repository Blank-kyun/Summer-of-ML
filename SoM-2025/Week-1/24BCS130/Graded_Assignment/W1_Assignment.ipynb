### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
ans=Pokemon[(Pokemon["Legendary"]==False) & (Pokemon["Type 1"]=="Ghost") & (Pokemon["Generation"]==4)]
ans=ans[ans["Speed"]==ans["Speed"].min()]
print(ans)
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
ans=Pokemon[(Pokemon["Legendary"]==False) & (Pokemon["Attack"]<Pokemon.loc["Charizard","Attack"]) & (Pokemon["Defense"]>Pokemon.loc["Charizard","Defense"])]
print(len(ans))
```

### Which type of legendary pokemons are the most common?

```python
#your code here
ans=Pokemon[Pokemon["Legendary"]==True]
ans=ans["Type 1"].value_counts()
print(ans.mode()[0])
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
ans=Pokemon[(Pokemon["Legendary"]==False) & (Pokemon["Generation"]==3)]
ans1=Pokemon[(Pokemon["Legendary"]==True) & (Pokemon["Generation"]==6)]
ans=ans[ans["HP"]>ans1["HP"].min()]
print(len(ans))
```
