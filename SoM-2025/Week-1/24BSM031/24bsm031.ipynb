### Which type of legendary pokemons are the most common?

```python
#your code here
# Which type of legendary pokemons are the most common?
legendary = df[df['Legendary'] == True]
most_common = legendary['Type 1'].value_counts().idxmax()
print(f"The most common type of legendary pokemons is:  "+most_common)
```

```python
gen3 = df[(df['Generation'] == 3) & (df['Legendary'] == False)]
gen6 = df[(df['Generation'] == 6) & (df['Legendary'] == True)]
weakest_gen6 = gen6['HP'].min()
gen3[gen3['HP'] > weakest_gen6].shape[0]
```
