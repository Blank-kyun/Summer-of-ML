```python

df1.sort_values(by='Generosity',ascending=False,inplace=True)
df1[df1['Regional indicator']=='East Asia'].head(5)

```

```python
tp=df1[df1['Regional indicator']=='East Asia'].head(5)
sns.lineplot(x=tp.index,y=tp['Ladder score'],color='blue',marker='o',markersize=10)
plt.xlabel('Countries')
plt.ylabel('Ladder score')
plt.title('Ladder score of top 5 countries in East Asia')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
sns.set()
```

```python
df1_2021=df1.loc[['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']]
plt.title("Countries vs their 'Logged GDP per capita'")
plt.xlabel('Countries')
plt.ylabel('Logged GDP per capita')
plt.scatter(df1_2021.index, df1_2021['Logged GDP per capita'], marker='o')
plt.xticks(df1_2021.index)
plt.show()
```
