# Task 2 : Visualization of results using Matplotlib library


Display the data of top 5 east asian countries based on 'Generosity'


```python
import matplotlib.pyplot as plt
```

```python
east_asia = df1[df1['Regional indicator']=='East Asia']
east_asia_req = east_asia.sort_values(by='Generosity',ascending=False).head(5)
```

Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)


```python
plt.figure(figsize=(10,8))
plt.plot(east_asia_req['Country name'],east_asia_req['Ladder score'])
plt.title("Variation of Ladder Score among Top 5 East Asian Countries (by Generosity)")
plt.xlabel("Country Name")
plt.ylabel("Ladder Score")
plt.show()
```

Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'


```python
df_2021 =df1[df1['Country name'].isin(['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'])]
plt.figure(figsize=(10,8))
plt.scatter(df_2021['Country name'],df_2021['Logged GDP per capita'])
plt.xlabel("Country")
plt.ylabel("Logged GDP per capita")
plt.show()
```
