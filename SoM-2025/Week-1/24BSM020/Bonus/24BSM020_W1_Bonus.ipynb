### Task-2

```python
east_asian_countries = df1[(df1['Regional indicator']=='East Asia')]
east_asian_countries.sort_values(by='Generosity', ascending=False, inplace=True)
east_asian_countries.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(data=east_asian_countries, x='Country name', y='Generosity', marker='o')
plt.xticks(rotation=45)
```

```python
countries=['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021=pd.DataFrame(df1,columns=df1.columns,)
df_2021=df_2021[df_2021['Country name'].isin(countries)]
df_2021
```

```python
sns.scatterplot(data=df_2021, x='Logged GDP per capita', y='Country name', hue='Country name', style='Country name', s=100)
sns.set(style='whitegrid')
plt.title('Logged GDP per capita of South Asian Countries in 2021')
plt.xlabel('Logged GDP per capita')
plt.ylabel('Country Name')
plt.legend(title='Country Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
