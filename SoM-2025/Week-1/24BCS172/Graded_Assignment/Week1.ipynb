### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
#your code here
sns.scatterplot(x='Attack', y='Defense', hue='Legendary', data=data)
plt.title("Attack vs Defense - Legendary and Others")


#Most Pokémon have a balance between Attack and Defense, but Legendary Pokémon often appear in the top-right, showing high values in both.
```

```python
data['Type 1'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title("Top 10 Most Common Pokémon Types")
#Most common typpe of pokemon is Water type
```

```python
stats=data.groupby('Legendary')['Total'].mean().reset_index()
sns.barplot(data=stats, x='Legendary', y='Total')
plt.title("Average Total Stats: Legendary vs Non-Legendary Pokémon")
plt.ylabel("Average Total Stats")
plt.xlabel("Legendary Status")
plt.xticks([0, 1], ['False (Normal)', 'True (Legendary)'])
plt.show()
# From this bargraph, we can tell that legendary pokemons in general have a lot higher total stats compared to non-legendary pokemons.
```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
#your code here
temp=data[(data.Legendary==False) & ((data['Type 1']=='Ghost') | (data['Type 2']=='Ghost')) & (data.Generation==4)]
temp2=temp.sort_values(by='Speed',ascending=False)
print(temp2.iloc[1][['Name']])
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
temp=data[data.Name=="Charizard"]
temp2=data[(data['Defense']>temp['Defense'].values[0]) & (data.Legendary==False) & (data['Attack']<temp['Attack'].values[0])].Name.count()
print(temp2)
```
