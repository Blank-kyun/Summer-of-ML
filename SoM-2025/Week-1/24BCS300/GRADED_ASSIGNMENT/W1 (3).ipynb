### 1. How many pokemons have 'Mega' in their name?

```python
import pandas as pd
df = pd.read_csv('W1_Pokemon.csv')


mega_pokemon = df[df["Name"].str.contains("Mega", case=False, na=False)]


print("Number of Pokémon with 'Mega' in their name:", mega_pokemon.shape[0])

```

### 2. What is the standard deviation of Sp. Def. in the dataset ?

```python
import pandas as pd


df = pd.read_csv('W1_Pokemon.csv')


std_sp_def = df['Sp. Def'].std()


print("Standard deviation of Sp. Def.:", std_sp_def)

```

### 3. What percentage (upto 3 decimal places) of pokemons are legendary ?

```python
import pandas as pd

df = pd.read_csv('W1_Pokemon.csv')


total_pokemon = len(df)


legendary_count = df[df["Legendary"] == True].shape[0]

legendary_percentage = (legendary_count / total_pokemon) * 100

print(f"Percentage of Legendary Pokémon: {legendary_percentage:.3f}%")


```

### 4. Name the pokemon(s) with Maximum Defense.

```python
import pandas as pd


df = pd.read_csv('W1_Pokemon.csv')


max_defense = df["Defense"].max()


strongest_pokemon = df[df["Defense"] == max_defense][["Name", "Defense"]]


print("Pokémon(s) with Maximum Defense:\n", strongest_pokemon)


```

### 5. Which poison pokemon has the strongest attack ?

```python
import pandas as pd

df = pd.read_csv('W1_Pokemon.csv')


poison_pokemon = df[df["Type 1"] == "Poison"]


max_attack = poison_pokemon["Attack"].max()

strongest_poison_pokemon = poison_pokemon[poison_pokemon["Attack"] == max_attack][["Name", "Attack"]]

print("Poison Pokémon with the strongest Attack:\n", strongest_poison_pokemon)



```

### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('W1_Pokemon.csv')


type_totals = df.groupby('Type 1')['Total'].mean().reset_index()


type_totals = type_totals.sort_values('Total', ascending=False)


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
ax = sns.barplot(data=type_totals, x='Type 1', y='Total', palette='Set2')


plt.title('Average Total Stats by Primary Pokémon Type')
plt.xlabel('Primary Type')
plt.ylabel('Average Total Stat')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('W1_Pokemon.csv')


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=df,
    x='Speed',
    y='Total',
    hue='Legendary',
    palette='Set1',
    alpha=0.7
)


plt.title('Pokémon: Total Stats vs. Speed')
plt.xlabel('Speed')
plt.ylabel('Total Stats')
plt.legend(title='Legendary')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('W1_Pokemon.csv')


gen_avg = df.groupby('Generation')['Total'].mean().reset_index()


sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.lineplot(data=gen_avg, x='Generation', y='Total', marker='o', color='mediumseagreen')

plt.title('Average Total Stats by Pokémon Generation')
plt.xlabel('Generation')
plt.ylabel('Average Total Stats')
plt.xticks(gen_avg['Generation'])
plt.tight_layout()
plt.show()

```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
import pandas as pd


df = pd.read_csv("W1_Pokemon.csv")


gen4_pokemon = df[df["Generation"] == 4]


ghost_pokemon = gen4_pokemon[(gen4_pokemon["Type 1"] == "Ghost") | (gen4_pokemon["Type 2"] == "Ghost")]
non_legendary_ghosts = ghost_pokemon[ghost_pokemon["Legendary"] == False]


sorted_ghosts = non_legendary_ghosts.sort_values(by="Speed", ascending=False)


if sorted_ghosts.shape[0] > 1:
    second_fastest_ghost = sorted_ghosts.iloc[1]
    print("Second fastest non-legendary Ghost-type Pokémon from Generation 4:")
    print(second_fastest_ghost[["Name", "Speed"]])
else:
    print("Not enough non-legendary Ghost-type Pokémon in Generation 4 to find the second fastest")

```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
import pandas as pd

df = pd.read_csv("W1_Pokemon.csv")


non_legendary_pokemon = df[df["Legendary"] == False]

filtered_pokemon = non_legendary_pokemon[(non_legendary_pokemon["Defense"] >78) & (non_legendary_pokemon["Attack"] < 84)]

count_pokemon = filtered_pokemon.shape[0]


print("Number of non-legendary Pokémon with stronger Defense but weaker Attack than Charizard:", count_pokemon)

```
