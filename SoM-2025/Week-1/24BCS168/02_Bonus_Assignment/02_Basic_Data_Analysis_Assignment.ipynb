### 12) Summarize only the occupation column

```python
print("Data type:", ds['occupation'].dtype)
print("Number of unique occupations:", ds['occupation'].nunique())
print("Unique occupations:", ds['occupation'].unique())
print("Most frequent occupation:", ds['occupation'].mode()[0])
print("Occupation value counts:")
print(ds['occupation'].value_counts())
```

### 13) What is the mean age of users?

### 14) What is the age with least occurrence?

```python
count = ds['age'].value_counts()
ages = count[count==count.min()]
print(', '.join(str(age) for age in ages.index))
```
