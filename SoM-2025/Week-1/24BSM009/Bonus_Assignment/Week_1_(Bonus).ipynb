# **TASK 2**

**Display the data of top 5 east asian countries based on 'Generosity'**

```
east_asia_genero = df1[df1['Regional indicator'] == 'East Asia'].sort_values('Generosity' , ascending = False)
east_asia_genero.index =range(1, len(east_asia_genero) + 1)
result = east_asia_genero.head(5)
plt.bar(result['Country name'] , result['Generosity'])
plt.xticks(rotation = -90)
plt.title("Top 5 east asian countries based on 'Generosity'")
plt.show()
```

**Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)**

```
plt.figure(figsize=(7,5))
plt.plot(result['Ladder score'])
plt.xticks(result.index, result['Country name'], rotation = -90)
plt.xlabel('Country Name')
plt.ylabel('Ladder Score')
plt.title('Variation of Ladder Score among East Asian Countries')
plt.show()
```

**Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'**

```
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(countries)]
relatn = df_2021['Explained by: Log GDP per capita']
plt.scatter(df_2021['Country name'] , relatn)
plt.xticks(rotation = 45)
plt.title("Countries vs their Logged GDP per capita")
plt.show()
```
