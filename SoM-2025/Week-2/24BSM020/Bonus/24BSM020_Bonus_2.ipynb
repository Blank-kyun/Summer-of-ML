```python
# This means we only have valid 7437 cf_ratings and 10280 cc_ratings,let us find out where we have both the ratings
df_1=df.dropna()
print(df_1.shape)
y=df_1['cf_rating']
x=df_1['cc_rating']
```

```python
plt.grid()
plt.scatter(x,y,color='blue')
plt.xlabel('CodeChef Rating')
plt.ylabel('CodeForces Rating')
plt.show()
```

```python
sns.regplot(x=x,y=y,scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.grid()
plt.xlabel('CodeChef Rating')
plt.ylabel('CodeForces Rating')
plt.title('Regression Plot')
plt.show()
```

```python
def gradient_descent(m_now,b_now,points,L):
    m_gradient,b_gradient=0,0
    n=len(points)
    # We will use the partial derivative to calculate the gradient
    for i in range(n):
        x=points.iloc[i].cc_rating
        y=points.iloc[i].cf_rating
        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
        b_gradient += -(2/n) * (y - (m_now * x + b_now))
    
    m=m_now-(L*m_gradient)
    b=b_now-(L*b_gradient)
    return m,b

m=0
b=0
L=0.001
epochs=100

for i in range(epochs):
    m,b=gradient_descent(m,b,df_1,L)
```

```python
print('m :',m,"\n","b :",b)
```
