```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://drive.usercontent.google.com/u/0/uc?id=1VfCaU5vFVWsSYrvKQF2x9iS6I7KoWEVH&export=download")
df.shape
```

```python
y=df['cf_rating']
x=df['cc_rating']
print('Values cc :',len(x),'Values cf :',len(y))
x=x.dropna()
y=y.dropna()
print("Removing nan cc :",len(x),"Removing nan cf :",len(y))
```

```python
df_1=df.dropna()
print(df_1.shape)
y=df_1['cf_rating']
x=df_1['cc_rating']
```

```python
plt.scatter(x,y,color='green')
plt.title("Scatter Plot between CC and CF")
plt.xlabel('CodeChef rating')
plt.ylabel('CodeForces rating')
plt.show()
```

```python
sns.regplot(x=x,y=y,scatter_kws={"color": "green"}, line_kws={"color": "red"})
plt.title('Regression plot')
plt.show()
```

```python

def estimate_coef(x, y):

    n = np.size(x)

    m_x = np.mean(x)
    m_y = np.mean(y)

    SS_xy = np.sum(y*x) - n*m_y*m_x
    SS_xx = np.sum(x*x) - n*m_x*m_x

    b_1 = SS_xy / SS_xx
    b_0 = m_y - b_1*m_x

    return (b_0, b_1)

def plot_regression_line(x, y, b):
    plt.scatter(x, y, color = "m",marker = "o", s = 30)

    y_pred = b[0] + b[1]*x
    plt.plot(x, y_pred, color = "g")
    plt.xlabel('CodeChef rating')
    plt.ylabel('CodeForces')
    plt.show()

b = estimate_coef(x, y)
print("Estimated coefficients:")
print("Constant b_0 =",b[0])
print("Slope of graph b_1 =",b[1])


plot_regression_line(x, y, b)
```
