```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://drive.usercontent.google.com/u/0/uc?id=1VfCaU5vFVWsSYrvKQF2x9iS6I7KoWEVH&export=download')
df.shape
```

```python
df=df.dropna()
df
```

### Plotting Scatter Plot of Dataset

```python
plt.scatter(x='cc_rating',y='cf_rating',data=df)
plt.title("Scatter Plot between CodeChef and CodeForces", fontsize=15)
plt.xlabel('CodeChef rating', fontsize=15)
plt.ylabel('CodeForces rating' ,fontsize=15)
plt.show()
```

### Plotting Output First Through sns.regplot

```python
sns.regplot(x='cc_rating',y='cf_rating',data=df, line_kws={"color": "r"})
plt.title('Regression plot',fontsize=15)
plt.show()
```

### Performing Linear Regression Without Scikit Learn 

```python
x=df.cc_rating
y=df.cf_rating

n = np.size(x) # no of samples

# mean of samples
m_x = np.mean(x)    
m_y = np.mean(y)

SS_xy = np.sum(y*x) - n*(m_y)*(m_x) # SS_xy is the sum of cross-deviations of y and x
SS_xx = np.sum(x*x) - n*(m_x)*(m_x) # SS_xx is the sum of squared deviations of x

b_1 = (SS_xy) / (SS_xx) # calculating coefficient of x (slope of regression line)
b_0 = m_y - (b_1)*(m_x) # calculating constant term (y intercept of regression line)

plt.scatter(x, y) # scatter plot of dataset

y_pred = b_0 + (b_1*x) # equation of regression line

plt.plot(x, y_pred, color = "r") # plotting regression line
plt.xlabel('CodeChef rating',fontsize=10)
plt.ylabel('CodeForces',fontsize=10)
plt.title('Regression Plot',fontsize=15)
plt.show()
```

```python
print(f"Slope of Line: {b_1} And Y intercept of line is: {b_0}")
```
